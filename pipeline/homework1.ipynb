### Q5

SELECT 
    gtt."PULocationID", tzl."Zone",
    COUNT(*) AS "NumberOfTrips"
FROM 
    public.green_taxi_trips gtt
INNER JOIN 
    public.taxi_zone_lookup tzl 
    ON gtt."PULocationID" = tzl."LocationID"
WHERE 
    DATE(gtt.lpep_pickup_datetime) = '2025-11-18'
GROUP BY 
    gtt."PULocationID",tzl."Zone"
ORDER BY 
    "NumberOfTrips" DESC;

Answer: East Harlem North

### Q6

SELECT 
    tzl1."Zone" AS "DropOff_Zone",
    MAX(gtt."tip_amount") AS "LargestTip"
FROM 
    public.green_taxi_trips gtt
INNER JOIN 
    public.taxi_zone_lookup tzl 
    ON gtt."PULocationID" = tzl."LocationID"
INNER JOIN 
    public.taxi_zone_lookup tzl1 
    ON gtt."DOLocationID" = tzl1."LocationID"
WHERE 
    DATE(gtt.lpep_pickup_datetime) BETWEEN '2025-11-01' AND '2025-11-30'
    AND tzl."Zone" = 'East Harlem North'
GROUP BY 
    tzl1."Zone"
ORDER BY 
    "LargestTip" DESC

Answer: 
Yorkville West
